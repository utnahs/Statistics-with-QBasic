```python
# Q1. Second Highest Value
numbers = [12, 45, 7, 34, 45, 23]   # Easy way sorted(numbers)[-2]
def second_largest(List):
    temp=0
    Length=len(List)
    for i in range(Length-1):
        for j in range(Length-1-i):
            if List[j]>List[j+1]:
                temp=List[j]
                List[j]=List[j+1]
                List[j+1]=temp
    return List[-2] #gives the second largest
second_largest(numbers)
```

```python
# Q2. Remove Duplicates (Preserve Order)
values = [1, 2, 2, 3, 4, 3, 5, 1]
def remove_duplicates(list_):
    new_list=[]
    for i in range(len(values)):
        c=0
        for j in range(len(new_list)):
            if values[i]==new_list[j]:
                c+=1
            if c>1:
                break
        if c<1:
            new_list.append(list_[i])
    return new_list
remove_duplicates(values)
```

```python
# Q7. Rotate List
# Rotate the list to the right by k positions.
items = [1, 2, 3, 4, 5]
k = 2                               # k = k % len(items)    Use modulo to handle cases where n_steps is larger than list length
print(items)
while k>=1:
    last=items.pop(-1)              # pop the last element
    items.insert(0,last)            # insert the removed elemnt to the first to shift by one place do the same to shift by two place
    print(items)
    k-=1


```

```python
# Q15. Merge Dictionaries (Sum Values)
d1 = {"A": 100, "B": 200}
d2 = {"B": 150, "C": 300}

merged=d1.copy()
for key in d2:
    if key in merged:
        merged[key]+=d2[key]
    else:
        merged[key]=d2[key]
print(merged)
```

```python
# Q17. Reverse Dictionary
# Reverse it so grades become keys.
grades = {"Alice": "A", "Bob": "B", "Charlie": "A"}

# {y: x for x,y in grades.items()} note: works but removes duplicates

grades2={}
for key,value in grades.items():
    grades2.setdefault(value,[]).append(key)
print(grades2)
```

```python
# Q25. Date Format Conversion
# Convert to YYYY/MM/DD.
date = "15-08-2024"
day,month,year=date.split("-")
print(f"{year}/{month}/{day}")

i=0
parts=[]
while i<len(date):
    ch=''
    while i<len(date) and date[i]!="-":
        ch+=date[i]
        i+=1
    parts.append(ch)
    i+=1
date=parts[0]
month=parts[1]
year=parts[2]
print(year,month,date)

```
